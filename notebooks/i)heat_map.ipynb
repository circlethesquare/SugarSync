```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/clean/df_processed.csv")
```

```python
#creating a DF for my results to create a heatmap
#it is used in my tableau presentation 
```

```python

# Step 1: Create a dictionary with the data
data = {
    'KNN': [0.4971751412429379],
    'DecisionTreeClassifier': [0.4406779661016949],
    'RandomForestClassifier': [0.5706214689265536]
}

df = pd.DataFrame(data, index=['full_data'])

# Step 3: Add the second row
df.loc['readmission'] = [0.5875706214689266, 0.5706214689265536, 0.615819209039548]

# Step 4: Add the third row
df.loc['sample_data'] = [0.31711145996860285, 0.40188383045525905, 0.4489795918367347]

# Step 5: Add the last row
df.loc['extra_readmission'] = [0.5645933014354066, 0.5119617224880383, 0.6028708133971292]

# Display the DataFrame
print(df)
```

```python
df.to_csv(path_or_buf="../data/clean/results.csv",header=0)
```

```python
sns.heatmap(df)
```

```python
plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(df, annot=True, cmap="YlGnBu", fmt=".2f")

# Rotate x-axis labels
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=0)
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0)
plt.savefig('../graphs/results_heat_map.png')
plt.show()
```
